## 使用Urllib获取网页源代码

使用urllib库可以很方便的给web服务器发送请求，并获取到相应的网页响应内容。

下面是一个简单的示例，我们通过urllib去获取百度首页的网页内容：

````python
from urllib.request import urlopen

response = urlopen(url="http://www.baidu.com")

print(response.read().decode())
````

关于urllib.request.urlopen参数的介绍：

```python
urllib.request.urlopen(url, data=None, [timeout, ]*, cafile=None, capath=None, cadefault=False, context=None)
```

- url:

    发送请求的网址，它可以是字符串或**请求对象**。

- data: 

    是一个指定要发送到服务器的附加数据的对象，或者如果不需要此类数据，则为“None”。

- timeout:

    超时参数，是可选的。指定阻塞操作（如连接尝试）的超时（以秒为单位）（如果未指定，将使用全局默认超时设置）。这实际上只适用于HTTP、HTTPS和FTP连接。

前面的三个参数是最常用的。下面的几个参数都和https（超文本传输安全协议）有关。

- contenxt: 

    上下文,如果指定了上下文，则它必须是描述各种SSL选项的ssl.ssl context实例。

- cafile和capath：

    这两个参数都是可选的，为https请求指定一组受信任的CA证书。ca file应该指向包含CA证书束的单个文件，而capath应该指向哈希证书文件的目录。

- cadefault
    
    参数经常被忽略。

## 请求对象 Request Object

可以将url先构造成一个Request对象，传进urlopen，Request对象存在的意义是便于在请求的时候传入一些信息。

### 构造方法
```python
urllib.request.Request(url, data=None, headers={}, origin_req_host=None, unverifiable=False, method=None)
```
**参数：**
- url:
    
    发送请求的网址，它必须是字符串。

- data: 

    是一个指定要发送到服务器的附加数据的对象，或者如果不需要此类数据，则为“None”。对于HTTP POST请求方法，数据应该是/x-www-form-urlencoded格式。urllib.parse.urlencode（）函数接受2个元组的映射或序列，并返回此格式的ASCII字符串。在用作数据参数之前，应该将其编码为字节。

- header：

    应该是一个字典，并且将被视为每个键和值作为参数调用了add_header（）。这通常用于“欺骗”User-Agent值，浏览器使用该值来标识自身——一些HTTP服务器只允许来自普通浏览器的请求，而不允许来自脚本的请求。例如，mozilla firefox可以将自己标识为“mozilla/5.0（x11；u；linux i686）gecko/20071127 firefox/2.0.0.11”，而urlib的默认用户代理字符串是“python urlib/3.6” 如果存在数据参数，则应包含适当的Content-Type头。如果未提供此标题，并且数据不是“无”，则将默认添加Content-Type类型：application/x-www-form-urlencoded。最后两个参数只对正确处理第三方HTTP cookie感兴趣：

- origin_req_host：

    是源站事务的请求主机，如RFC2965所定义。它默认为http.cookiejar.request_host（self）。这是由用户启动的原始请求的主机名或IP地址。

- unverifiable：
        
    指出请求是否是不可验证的，如RFC2965所定义的。它默认为False。无法验证的请求是用户没有可供批准的URL的请求。例如，如果请求是针对HTML文档中的图像，并且用户没有批准自动提取图像的选项，那么这应该是True。

- method:

    是一个字符串，指示将使用的HTTP请求方法（例如“head”）。如果提供，则其值存储在method属性中，并由get_method（）使用。如果数据为“无”，则默认值为“get”，否则为“post”。

````python
from urllib import request,parse

url='http://www.hao123.com'

headers={

    'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/55.0.2883.95 Safari/537.36'

}

d ={ 'name':'admin'}

data=bytes(parse.urlencode(d),encoding='utf-8')

#利用Request将headers，dict，data整合成一个对象传入urlopen

req = request.Request(url,data,headers,method='POST')

response=request.urlopen(req) 

print(response.read().decode('utf-8'))
````

## 使用Urllib发送GET请求

### GET请求的参数

比如说，我们用百度搜索“爬虫”，地址栏会显示：

https://www.baidu.com/s?wd=爬虫

但是服务器接收到的却是

https://www.baidu.com/s?ie=UTF-8&wd=%E7%88%AC%E8%99%AB

原因是在浏览器/客户端在向服务器传递GET请求参数的过程中：

- 数据如果是英文字母/数字，原样发送， 

- 如果是空格，转换为+，

- 如果是中文/其他字符，则直接把字符串用BASE64加密，及“%”加上“字符串的16进制ASCII码”。

另外：

GET提交的数据最大是2k,原则上url长度无限制，限制实际上取决于浏览器，大多数浏览器通常都会限制url长度在2K个字节

### 参数的编码和解码

GET请求提交数据，数据需要编码成 URL编码格式，然后做为url的一部分，或者作为参数传到Request对象中。



````python
from urllib import parse

word = {"wd" : "爬虫"}
````

**通过parse.urlencode()方法，将字典键值对按URL编码转换，从而能被web服务器接受。**

````python
result=parse.urlencode(word)  

print(result)
````

**通过parse.unquote()方法，把 URL编码字符串，转换回原先字符串。**

````python
print(parse.unquote(result))
````

### 发送携带参数的GET请求

````python
from urllib import request,parse



url = "http://www.baidu.com"

word = {

"wd":"爬虫",

}

word = parse.urlencode(word)

newurl = url+"?"+word

headers = {

"User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Safari/537.36",

}

req = request.Request(url=newurl,headers=headers)

response = request.urlopen(req)



print(response.read())
````

## 使用Urllib发送POST请求

### POST请求的参数

get和post都是HTTP与服务器交互的方式，post把数据放在HTTP的包体内（requrest body）。post提交的数据大小理论上没有限制。

前面我们讲得Request请求对象中有一个data参数，它就是用在post请求的，我们要传送的数据就是这个参数，data是一个字典，将要发送的数据以键值对的形式放在里面就可以了。

### POST请求实例

豆瓣电影 Top 250，即历史上最著名的250部电影列表，提供了一种获取数据的post请求方式

https://movie.douban.com/j/chart/top_list

大家可以将下面的数据用post请求发送到这个地址：

formdata = {

​    'type':'11',         #类型

​    'interval_id':'100:90',

​    'action':'',

​    'start':'0',          #开始序号

​    'limit':'10'          #每次返回的记录数

}

完整的代码如下：

````python
import json

from urllib import request,parse



url = "https://movie.douban.com/j/chart/top_list?"

headers = {"User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Safari/537.36",}

formdata = {

"type":"11",

"interval_id":"100:90",

"action":" ",

"start":"0",

"limit":"10",

}

data = parse.urlencode(formdata).encode()

req = request.Request(url=url,data=data,headers=headers)

response = request.urlopen(req)



print(json.loads(response.read()))
````
